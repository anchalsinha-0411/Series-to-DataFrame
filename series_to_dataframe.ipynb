```python
d1=pd.Series({'Material':'Wooden','Colour':"Maroon",'Price':25000})
d2=pd.Series({'Material':'Plywood','Colour':"Yellow",'Price':20000})
d3=pd.Series({'Material':'Plastic','Colour':"Red",'Price':1500})
d4=pd.Series({'Material':'Stainless Steel','Colour':"Silver",'Price':55000})
d5=pd.Series({'Material':'Wooden','Colour':"Light Blue",'Price':2500})
d6=pd.Series({'Material':'Aluminium','Colour':"Golden",'Price':65000})
```

```python
df=pd.DataFrame([d1,d2,d3,d4,d5,d6],index=['Sofa','Dining Table','Chair','Sofa','Chair','Dining Table'])
print(df)
```

```python
df
```

```python
#display the details of the chair and sofa
df.loc[['Sofa','Chair']]
```

```python
#display the furniture details which price is more than 25000
df[df['Price']>25000]
```

```python
#display the furniture details price under 10000
df[df['Price']<10000]
```

```python
#creating a dataframe using the 2d dictionary to store the following records
import pandas as pd

data={
    'Activity1':{'Blue House':98,'Red House':87,'Green House':59,'Yellow House':78},
    'Activity2':{'Blue House':85,'Red House':76,'Green House':67,'Yellow House':99},
    'Activity3':{'Blue House':88,'Red House':80,'Green House':91,'Yellow House':55}
}
```
